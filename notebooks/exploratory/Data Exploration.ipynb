```python
import os
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv(os.path.join('../../data/raw', 'train.csv'))
test_data = pd.read_csv(os.path.join('../../data/raw', 'test.csv'))

train_data.set_index('id', inplace=True)
y_train = train_data['species']
train_data.drop('species', axis=1, inplace=True)
test_data.set_index('id', inplace=True)
```

### Manual Feature selection

```python
mask = ((train_data == 0).sum() / train_data.shape[0]) > 0.30
nb_useless = (mask).sum()
print("Number of features having more than 30% of their values equal to  0: ", nb_useless)

```

```python
train_data.loc[:, ~mask]
```

```python
train_data = train_data.loc[:, ~mask]
test_data = test_data.loc[:, ~mask]
```
